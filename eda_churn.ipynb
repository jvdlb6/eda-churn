## Transformação de Dado

```python
# Transformar coluna TotalCharges para Float - type one

df_contracts['TotalCharges'] = df_contracts['TotalCharges'].astype('float')
df_contracts.info()
```

```python
# Transformar coluna TotalCharges de String para Float - Type 2

df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'])
```

```python
# Transformar coluna TotalCharges de String para Float - Type 3

df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'], errors='coerce')
df_contracts.info()
```

Após a transformação da coluna TotalCharges do DataFrame Contracts, a coluna passou a ter 11 valores ausentes(null)

## Unificar Datagrames de Customers, Services e Contracts

```python
# Contar quantidade de registros
len(df_customers)
```

```python
len(df_services)
```

```python
len(df_contracts)
```

```python
# Renomear colunas
df_services.rename(columns={'customerID': 'IDCliente'}, inplace=True)
```

```python
df_services.info()
```

```python
# Unificar Dataframe de Custormers com Services, criando um terceiro DataFrame
df_temp = df_customers.merge(df_services, on=['IDCliente'])
```

```python
df_temp.info()
```

```python
# Unificar df_temp com Contracts, usando colunas de junção com nomes distintos
df_churn_temp = df_temp.merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'])
```

```python
df_churn_temp.info()
```

```python
# Unificar os três dataframes ao mesmo tempo, com colunas com nomes diferentes

df_churn = df_customers.merge(df_services, on=['IDCliente']).merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'])
```

```python
df_churn.info()
```

```python
# Removendo coluna de um Dataframe

df_churn.drop(['customerID'], axis=1, inplace=True)
```

```python
df_churn.info()
```

## Imputação de Valores Ausentes

```python
# Preenche os valores ausentes no dataframe com 0

df_churn.fillna(0)
```

```python
# Preenche valores conforme a coluna

df_churn.fillna(value={'TotalCharges': 0, 'Genero': 'Não Declarado'})
```

```python
# Preenche todos os valores ausentes de uma coluna com a média
media_TotalCharges = df_churn.TotalCharges.mean()
df_churn.fillna(value={'TotalCharges': media_TotalCharges})
```

> Lembrando que os valores não tiveram o metodo inplace pois iria modificar o dataset que usei como base, porém, para dados e uso real, é necessário o `inplace=True` para modificar o dataset
