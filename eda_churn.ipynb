## Análise Bivariada

```python
# Hipótese: Clientes com Contrato do tipo Month-to-Month mais propensos a cancelar o serviço (churn)
# Construir uma tabela de contingência entre as variáveis Contract e Churn - Quantidade de ocorrências

pd.crosstab(df_churn['Churn'], df_churn['Contract'], margins=True, margins_name='Total')
```

```python
# Construir uma tabela de contingência entre as variáveis Contract e Churn - Proportões

pd.crosstab(df_churn['Churn'], df_churn['Contract'], margins=True, normalize='index', margins_name='Total')
```

> Considerando o público total, os clientes de contrato mensal representam 55%. Porém, se considerarmos apenas o clientes que abandonaram o serviço, 88% tinham contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo do contrato mensal e o churn.

```python
# Avaliar a correlação entre duas variáveis categóricas(qualitativas)

# Executar um teste de hipótese chamado Chi-Square de independência ou Qui-quadrado de Pearson
# Num testes de hipótese, temos duas hipóteses:
# - H0: Hipótese Nula - Não existe correlação entre as variáveis
# - H1: Hipótese Alternativa - Existe correlação entre as variáveis

# O teste serve para confimar ou recusar a hipótese nula H0
# Quando a probabilidade de observarmos o H0 é inferior a 0.05 (p-value), recusamos H0 em favor de H1
```

```python

# Gerar um DF da Crosstab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn['Churn'], df_churn['Contract'])
```

```python
# Calcular os Scores e P-Values
from scipy.stats import chi2_contingency
```

```python
# Tupla com os resultados do teste
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar números com decimais sem a notação científica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar um DataFrame com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'P-Values': pvalues_churn_contract})
```

```python
# Analisar Scores e P-Values
df_chi_scores_churn_contract
```

> Conforme teste estatístico(Chi-Square), o P-value <= 0.05, desta forma rejeitamos a hipótese nula, ou seja, as variáveis não são independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação.

```python
# Avaliar a correalação entre uma variável qualiteativa e uma quantitativa
# Usar o mesmo teste Chi-Square mas transformar a vairável quantitativa em qualitativa
# Hipóstese: Clientes com menos de 6 meses de contrato são mais propensos a churn

# Criar uma coluna nova no DataFrame com base numa condição de outra coluna
# Coluna = 'TempoMenor6Meses'

import numpy as np

df_churn['TempoMenor6Meses'] = np.where(df_churn['tenure'] < 6, 'Yes', 'No')
```

```python
df_churn[['tenure', 'TempoMenor6Meses']].head(10)
```

```python
# Gerar uma crosstab (sem totais)
df_crosstab_churn_tenure = pd.crosstab(df_churn['Churn'], df_churn['TempoMenor6Meses'])
```

```python
# Calcular os scores
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'P-Values': pvalues_churn_tenure})
```

```python
# Analisar os resultados
df_chi_scores_churn_tenure
```

> Como o p-value <= 0.05, rejeitamos a H0, ou seja, as variáveis não são independentes. E vale mencionar que com base do Qui2 resultante das 2 análises, a correlação entre Churn e Tempo de COntrato < 6 meses é menor do que a correlação entre Churn e Tipo de Contrato

```python
# Criar um dataframe somente com clientes do tipo Month to Month

df_churn_month = df_churn[df_churn['Contract'] == 'Month-to-month']
df_churn_month
```

```python
# Histograma - Checar visualmente se os dados seguem uma distribuição normal

df_churn_month['TotalCharges'].plot(kind='hist')
```

```python
# Método de Tukey - IQR (Distribuição não normal)
# IQR - Range Interquartil
# IQR = Q3 - Q1
q1_totalcharges = df_churn_month['TotalCharges'].quantile(0.25)
q3_totalcharges = df_churn_month['TotalCharges'].quantile(0.75)
IQR = q3_totalcharges - q1_totalcharges
IQR
```

```python
# Limites para detectar outliers
lower_limit_totalcharges = q1_totalcharges - (1.5 * IQR)
upper_limit_totalcharges = q3_totalcharges + (1.5 * IQR)
lower_limit_totalcharges, upper_limit_totalcharges
```

```python
df_churn_month[(df_churn_month['TotalCharges'] < lower_limit_totalcharges) | (df_churn_month['TotalCharges'] > upper_limit_totalcharges)]
```

```python
# Zscore (Distribuição Normal)
# Indica quantos desvios padrões um ponto específico está distante da média
# zscore = (x - média) / desvio padrão
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_month['TotalCharges']))
```
